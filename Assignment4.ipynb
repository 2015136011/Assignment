## [1번문제]

```python
s = ['pig ham', 'cat dog', 'ham bird', 'dog pig'] #리스트s를 정의
f = open('s.txt','w') #s.txt를 읽기 모드로 연다.
a="\n".join(s) #문자열a에 리스트s를 사이에 "\n"을 넣어 이어붙여 넣는다.
f.write(a) #문자열 a를 s.txt에 적는다.
f.close() #s.txt를 닫는다
f = open('s.txt', 'r') #s.txt를 읽기모드로 연다.
print(f.read())
```

리스트 s를 사이에 \n을 넣어 문자열 a에 넣어주고 그 값을 s.txt에 넣어준다.

#### 1) 파일 s.txt을 읽어서 각 라인에 있는 첫 번째 단어(문자열) 자체들을 기준으로 라인별 정렬후 파일 s1.txt에 그 결과를 기록하는 코드를 작성하시오. 

```python
s1=sorted(s) #s를 크기순서대로 배열하여 새로운 리스트 s1에 저장한다.
f = open('s1.txt','w') #그 이후로는 전과 같음
a1="\n".join(s1)
f.write(a1)
f.close()
f = open('s1.txt', 'r')
print(f.read())
```

sorted(s)를 통하여 순서대로 s1에 넣어주고 전과 같이 s1.txt에 넣어준다

#### 2) 파일 s.txt을 읽어서 각 라인에 있는 두 번째 단어(문자열) 자체들을 기준으로 라인별 정렬후 파일 s2.txt에 그 결과를 기록하는 코드를 작성하시오.

```python
def cmp(n):
    return n[4] #다섯번째 원소를 반환하는 함수 생성
s2=sorted(s,key=cmp) #두번째 원소의 크기 순서로 배열하여 새로운 리스트 s2에 저장한다.
f = open('s2.txt','w') #그 이후로는 전과 같음
a2="\n".join(s2)
f.write(a2)
f.close()
f = open('s2.txt', 'r')
a = f.read()
print(a)
```

함수 cmp(n)을 통하여 두번째 원소의 크기를 sorted(s,key=cmp)로 정렬하고 전과 같이 s2.txt에 넣어준다

#### 3)파일 s.txt을 읽어서 각 라인들에 있는 모든 단어들을 순차적으로 다시 나열하되 각 라인에 세 개의 단어들이 오도록 하여 s3.txt에 기록하는 코드를 작성하시오.

```python
m=" ".join(s) #리스트s를 문자열 m에 공백 단위로 넣어준다.
a=m.split() #문자열 m을 다시 공백 단위로 나누어 리스트 a로 나눈다.
s3="" #문자열 s3 생성
j=0
while(j<(len(a))):
    s3=s3+a[j]+" " #리스트a의 각 원소를 문자열 s3에 넣어주면서
    if((j+1)%3==0): #[3의배수]번째 원소가 나올때마다 \n을 넣어 문제에서 출력하고자 하는 방식으로 문자열을 만든다.
        s3=s3+"\n"
    j=j+1
f = open('s3.txt','w') #그 이후로는 전과 같음
a3="".join(s3)
f.write(a3)
f.close()
f = open('s3.txt', 'r')
a = f.read()
print(a)
```

```python
import hashlib

def password(password):
    hash_object = hashlib.sha1(password.encode())
    hex_dig = hash_object.hexdigest()
    return hex_dig #password를 암호화하는 함수, sha모듈이 없어서 대체

def IDMatch(id): #동일한 ID가 있는지 확인하는 함수
    f=open('access.txt') 
    a=f.readlines() #access.txt에서 라인별로 문자열을 받아와 리스트 a에 저장
    f.close()
    i=0
    st=[]
    ID=[]
    PW=[]
    while(i<len(a)):
        st=a[i].split("]") #a[i]라는 한 줄의 문자열로 []:[],[],[] 형태이므로 ']'단위로 나누어 lst에 저장
        ID.append(st[0][1:]) #리스트의 첫번째 원소'[id'에서 앞에 있는 '['을 제외한 문자를 ID에 저장
        PW.append(st[1][2:]) #리스트의 두번째 원소':[pw'에서 앞에 있는 '[:'을 제외한 문자를 PW에 저장
        i=i+1 #라인수만큼 반복
    j=0
    while(j<len(ID)):
        if(id==ID[j]): #입력한 id가 리스트 ID의 원소 중 하나라면
            return PW[j] #해당 ID[j]와 같은 라인에 있는 pw인 PW[j]를 반환
        j=j+1
    return 0 #아니면 0 반환

def PWMatch(pw):
    f=open('access.txt')
    a=f.readlines() #access.txt에서 라인별로 문자열을 받아와 리스트 a에 저장
    f.close()
    i=0
    st=[]
    PW=[]
    NAME=[]
    while(i<len(a)):
        st=a[i].split("]") #a[i]라는 한 줄의 문자열로 []:[],[],[] 형태이므로 ']'단위로 나누어 lst에 저장
        PW.append(st[1][2:]) #리스트의 두번째 원소'[:pw'에서 앞에 있는 '[:'을 제외한 문자를 PW에 저장
        NAME.append(st[2][2:]) #리스트의 세번째 원소'[:name'에서 앞에 있는 '[:'을 제외한 문자를 NAME에 저장
        i=i+1 #라인수만큼 반복
    j=0
    while(j<len(PW)):
        if(password(pw)==PW[j]): #입력한 pw가 리스트 PW의 원소 중 하나라면
            return NAME[j] #해당 PW[j]와 같은 라인에 있는 name인 NAME[j]를 반환
        j=j+1
    return 0 #아니면 0 반환

print("Welcome to Our Service\n1.Sign Up\n2.Sign In\n3.Quit")
a=int(input())
f = open('access.txt','a')
f.close() #access.txt 생성

if(a==1):
    while(1):
        ID=input("ID를 입력하시오 : ")
        if (IDMatch(ID)!=0): #IDMatch(ID)가 0이면 없는 아이디 0이 아니면 있는 아이디 이므로
            print('Sorry, the entered ID is already used.') #0이 아니면 해당 문장 출력 후 다시 입력
        else:
            break #0이면 중복이 아니므로 입력 완료
    P=input("Password를 입력하시오 : ")
    Name=input("이름을 입력하시오 : ")
    School=input("학교를 입력하시오 : ")
    s="["+ID+"]:["+password(P)+"],["+Name+"],["+School+"]\n"
    f = open('access.txt', 'a')
    f.write(s) #문자열 s를 파일에 기록
    f.close()
    
elif(a==2):
    while(1):
        ID=input("ID를 입력하시오 : ")
        if(IDMatch(ID)==0): #0이면 없는 ID이므로 해당문장 출력
            print("Sorry, you are not a registered member.")
        else:
            break #0이 아니면 입력 완료
    while(1):
        P=input("Password를 입력하시오 : ")
        if(password(P)!=(IDMatch(ID))): #password(P)가 해당 ID에 연결된 PW를 출력하는 IDMatch(ID)와 다르면
            print("Sorry, the entered password is not correct.") #해당 문장 출력
        else: #일치한다면 로그인 성공으로
            print('Hello ['+PWMatch(P)+']!') #해당 암호와 연결된 NAME을 출력하는 PWMatch를 통하여 이름 출력
            break
elif(a==3):
    exit(1) #a가 3이면 바로 프로그램을 끝내도록
print("\n====회원리스트====")
f = open('access.txt', 'r')
print(f.read())
f.close()
```

문제의 요구조건을 만족시키기 위해서 함수를 총 세개를 만들었다. 입력한 pw를 암호화한 password(P)함수, 일치하는 ID가 있는지 찾는 함수인 IDMatch(id), 그리고 일치하는 PW가 있는지 찾는 함수인 PWMatch(pw)함수이다. password(P)함수는 hashlib를 통하여 쉽게 암호화 하였으며 IDMatch함수는 텍스트 파일의 문자열을 라인별로 받아서 i번째줄에 있는 ID와 PW를 각각 ID[i]와 PW[i]에 넣는 작업을 하여 사용자가 입력한 id가 ID리스트에 없으면 0을 반환하고 있다면은 해당 ID[j]와 같은 줄에 있는 pw인 PW[j]를 반환한다. PW[j]를 반환한 이유는 로그인시에 해당하는 아이디에 맞는 패스워드를 입력할때 꼭 필요하기 때문이다. PWMatch함수는 IDMatch 함수에서 ID리스트를 PW리스트로, PW리스트를 NAME리스트로 바꾼것이다. 암호가 일치하면 NAME[j]를 반환하는 이유는 IDMatch와 같이 최종적으로 그 라인에 있는 이름을 출력해야 하기 때문이다. 이후 메인함수들은 문제에서 요구하는대로 if문을 사용하여 만들었다. 

## [3번 문제]

range() 함수와 유사한 frange() 함수를 다음 조건을 참고하여 만드시오.

1) frange()의 인자 구성은 다음과 같이 range와 동일하지만 각 인수들은 음수를 받지 않는다고 가정한다. 

range(stop)

range(start, stop[, step])

2) frange() 함수의 인자에 대한 기본 시작(start) 값은 0.0이고, 기본 단계(step) 값은 0.1이다.

3) frange 사용 예

3-1) frange(0.5)

[0.0, 0.1, 0.2, 0.3, 0.4]

3-2) frange(1.0, 2.0)

[1.0, 1.1, 1.2, 1.3, 1.4, 1.5, 1.6, 1.7, 1.8, 1.9]

3-3) frange(2.2, 4.0, 0.5)

[2.2, 2.7, 3.2, 3.7]


```python
def frange(*var):
    lst = []
    if (len(var) == 1):
        start = 0.0
        step = 0.1
        stop = var[0]
    elif (len(var) == 2):
        start = var[0]
        step = 0.1
        stop = var[1]
    elif (len(var) == 3):
        start = var[0]
        step = var[2]
        stop = var[1]
    while (start < stop):
        lst.append(start) #start를 lst에 입력
        start=float("%.1f"%(start+step)) #start에 step을 더해줌
    return lst #lst 출력
    
print(frange(0.5))
print(frange(1.0, 2.0))
print(frange(2.2, 4.0, 0.5))
```

frange() 함수는 인수가 1개일때는 start에 인수를 입력해주고 step과 stop은 문제에서 주어진대로 기본 값인 각각 0.0, 0.1을 넣어준다. 인수가 2일때는 start와 stop에 각각 첫번째 인수와 두번째 인수 값을 입력해주고 step은 그대로 기본값이다. 인수가 3일때는 start,stop,step에 각각 첫번째 인수 두번째 인수 세번째 인수를 넣어준다.

## [4번문제]
가변인수를 받는 함수 sum()을 다음과 같은 조건을 참고하여 구현하시오

sum() 사용 예

- sum()
- 0

- sum(1, 2)
- 3

- sum(1, 2, 3, 4, 5)
- 15

- sum(1, 5, 7, 2, -10)
- 5

```python
def sum(*var):
    s=0 #더하는 변수 s를 초기화
    i=0 #i를 0으로 초기화
    while(i!=len(var)): #i가 가변인자의 갯수만큼
        s=s+var[i] #s에 i번째 가변인자를 더한다
        i=i+1 #i를 추가한다.
    return s #더한 값 반환

print(sum())
print(sum(1, 2))
print(sum(1, 2, 3, 4, 5))
print(sum(1, 5, 7, 2, -10))
```

처음에 i와 s를 0으로 초기화를 한다. i가 인수의 숫자가 아니라면 s에 첫번째 인수를 더해준다. i에 1을 더해주고 i가 인수의 숫자와 같아지면 여태까지 인수들을 더한 s값을 반환한다.

## [5번 문제]

여러 단어로 이루어진 문자열을 입력받아 각 단어의 첫글자로 이루어진 단어를 대문자로 출력하는 myinitial() 함수를 다음 조건을 참고하여 작성하시오.

1) 다음에 제시되는 함수들을 모두 이용해야 한다.

split

map

join

2) myinitial() 함수 사용 예

myinitial("as soon as possible")

ASAP

```python
def f(s):
    return s[0].upper() #첫번째 원소의 대문자값 반환
def myinitial(T):
    a=T.split() #입력한 T를 공백 단위로 나누어 리스트에 넣어준다.
    Y=list(map(f,a)) #리스트a를 map에 있는 f를 통하여 리스트에 각각 첫번째 원소의 대문자 처리한 것을 넣어준다.
    a1="".join(Y) #문자열a1에 리스트Y를 넣어준다.
    return a1 #a1 반환

T=str(input())
print(myinitial(T))
```

첫번째 원소의 대문자값을 반환하는 함수 f(s)를 정의 후에 myinitial(T) 함수에서 입력한 문자열 T를 공백 단위로 나누어 리스트에 넣어준다. 맵 함수를 통하여 리스트 a에 있는 원소를 f(a)에 거쳐서 반환값을 Y에 저장하고 Y에 있는 원소들을 a1이라는 문자열로 모두 이어붙혀서 출력한다. 

## [6번 문제]

음이 아닌 정수 n를 입력받으면 n! (factorial)을 계산하는 myfact() 함수를 재귀적 함수로 구현하시오.


[옵션] 참을 수 있는 정도 만큼의 수행시간을 직접 기다려보면서 n을 늘려보도록 합시다.


```python
def myfact(n):
    if(n<=1):
        return 1; #마지막 myfact(1) 은 1
    return n*myfact(n-1); #재귀적 함수를 위해서 이전 (n-1)의 함수값을 n과 곱한다.
print(myfact(2965))
```

n부터 차근차근 그 전의 팩토리얼값인 myfact(n-1)을 n에 곱하여 순환적으로 만들수 있다. 숫자를 넣어본 결과 2965까지는 출력이 되지만 2966부터 범위를 초과해서 그런지 RecursionError: maximum recursion depth exceeded in comparison 이라는 오류가 뜬다

```python
import string
#<script>제거
def BracketExcept(source1):
    k=source1.split() #소스 문자열을 공백 단위로 나누어 리스트에 넣는다.
    j=0
    a=''
    while (j < len(k)):
        if(k[j] == '<script>'): #리스트에 원소가 <script>나 <style>일시에 </script></style>이전까지 모든 원소를 무시한다.
            j=j+1
            while(k[j] != '</script>'):
                j=j+1
            a=a+''
            j=j+1
        elif(k[j]=='<style>'):
            j=j+1
            while(k[j] != '</style>'):
                j=j+1
            a=a+''
            j=j+1
        else: #원소가 <script>나 <style>이 아닐시에 a문자열에 넣어준다.
            a=a+k[j]+' '
            j=j+1 
    b="" #<>제외한 문자열을 넣어줄 a를 정의
    i=0 #i를 0으로 정의
    while (i < len(a)): #소스크기 만큼 반복하는 반복문
        if(a[i] == '<'): #요소가 <이 나올 경우
            i=i+1 #우선 다음 요소로 넘어가서
            while(a[i] != '>'): #요소가 >이 나올때까지
                i=i+1 #계속해서 다음 요소로 넘어간다.
            b=b+' ' # >이 나와서 while문을 탈출하면 공백을 a문자열에 넣어준다.
            i=i+1 # 다음 요소로 넘어감
        else:
            b=b+a[i] # 요소가 <가 아닐 경우에는 a문자열에 넣어준다.
            i=i+1 #다음 요소로 넘어감
    c=b.split() #공백 단위로 a의 문자열을 나누어 b에 넣어준다. (단어를 구분함)
    text=" ".join(c)
    return text

def replaceAll(text, dic):  ##문자열을 사전에 맵핑된 문자로 대체로 하는 함수
    for i,j in dic.items():
        text = text.replace(i,j,text.count(i)) #문자열에 사전에 키와 같은 문자가 존재하면 키에 맵핑된 value값으로 대체한다.
    return text   

def stripSpecialChar(text): ##특수문자 제거 함수
    keys = [ch for ch in string.punctuation] #구두문자를 key리스트로 만듬
    keys.append("●")
    keys.append("·")
    keys.append("★") #구두문자 이외의 특수문자를 key리스트에 추가
    keys.append("☆")
    keys.append("※")
    keys.append("nbsp")#html문자
    keys.append("lt")
    keys.append("amp")
    keys.append("quot")
    keys.append("middot")
    keys.append("uarr")
    keys.append("rarr")
    keys.append("darr")
    keys.append("harr")
    
    values = ['' for _ in keys]  #key리스트의 원소 수 만큼 ''문자를 원소를 갖는 value리스트를 만듬
    replaceDic = dict(zip(keys,values)) #key리스트 key로 value리스트를 value값으로 하는 사전 만듬

    return replaceAll(text,replaceDic) #추출한 순수 텍스트에서 사전에 맵핑된 제거할 문자들을 모두 제거
def wordOfFreq(l): #단어의 빈도수를 사전으로 반환하는 함수
    word = []   
    for w in l: #리스트의 각 원소들이
        if (w in word) == False : word.append(w) #word리스트에 포함된 단어가 아니면 단어를 리스트에 추가 (단어중복제거)          
    freg = [l.count(w) for w in l] #리스트에서 단어의 출현 빈도수를 리스트로 만듬
    
    return dict(zip(word,freg)) #단어와 단어의 빈도수로 맵핑된 사전 반환
def Abbreviation(source2,source3):
    k=source2.split() #소스 문자열을 공백 단위로 나누어 리스트에 넣는다.
    j=0
    a=''
    while (j < len(k)):
        if(k[j] == '<pre>#'): #<pre>이 나올 때부터
            j=j+1
            while(k[j] != '</pre>'): #</pre> 나오기 전까지의 원소들을 모두 a에 넣어준다.
                a=a+k[j]+'\n'
                j=j+1
        else:
            j=j+1
    b=a.split() #a의 문자열을 공백 단위로 나누어 리스트 b에 넣어준다.
    c=b[20:len(b)] #리스트의 20번째부터 슬라이싱을 하여 원하는 불용어를 넣어준다. 
    c=c+source3.split() #source3은 소스 코드 자체에 불용어가 다 있으므로 공백 단위로 나눈 값과 source2의 불용문을 합친다.
    return(c) #영어와 한글을 합친 불용문 출력

l1=stripSpecialChar(BracketExcept(source1))
l2=stripSpecialChar(BracketExcept(source4))
l3=stripSpecialChar(BracketExcept(source5))
l4=stripSpecialChar(BracketExcept(source6))
l5=stripSpecialChar(BracketExcept(source7))

#8-1
A=Abbreviation(source2,source3)
T = text.split()            #공백단위분리
U=[]
for i in range(0,len(T)):
    for j in range(0,len(A)):
        if (T[i] == A[j]):
            U.append(A[j])
def bulyongeo(l):
    values = ['' for _ in U]  #U리스트의 원소 수 만큼 ''문자를 원소를 갖는 value리스트를 만듬
    replaceDic = dict(zip(U,values)) #U리스트 U로 value리스트를 value값으로 하는 사전 만듬
    w=replaceAll(l,replaceDic) #추출한 순수 텍스트에서 사전에 맵핑된 제거할 문자들을 모두 제거
    W=w.split()
    return wordOfFreq(W)
B1=bulyongeo(l1)
B2=bulyongeo(l2)
B3=bulyongeo(l3)
B4=bulyongeo(l4)
B5=bulyongeo(l5)

import pickle
#8-2
f = open('cse.koreatech.ac.kr.html','w')
f.close()
f = open('www.koreatech.ac.kr.html','w')
f.close()
f = open('www.naver.com.html','w')
f.close()
f = open('www.daum.net.html','w')
f.close()
f = open('www.nytimes.com.html','w')
f.close()
#8-3
f = open('cse.koreatech.ac.kr.pickle.txt','wb')
pickle.dump(B1, f) 
f.close()
f = open("cse.koreatech.ac.kr.pickle.txt",'rb')     
z1 = pickle.load(f) 
f.close
f = open('www.koreatech.ac.kr.pickle.txt','wb')
pickle.dump(B2, f) 
f.close()
f = open("www.koreatech.ac.kr.pickle.txt",'rb')     
z2 = pickle.load(f) 
f.close
f = open('www.naver.com.pickle.txt','wb')
pickle.dump(B3, f) 
f.close()
f = open("www.naver.com.pickle.txt",'rb')     
z3 = pickle.load(f) 
f.close
f = open('www.daum.net.pickle.txt','wb')
pickle.dump(B4, f)
f.close()
f = open("www.daum.net.pickle.txt",'rb')     
z4 = pickle.load(f) 
f.close
f = open('www.nytimes.com.pickle.txt','wb')
pickle.dump(B5, f)
f.close()
f = open("www.nytimes.com.pickle.txt",'rb')     
z5 = pickle.load(f) 
f.close

def cmp_(a): #리스트 원소 중 두번째 원소를 반환함
    return a[1]
def Top3(z):
    Z = sorted(list(z.items()), key=cmp_,reverse = True)
    return Z

for i in range(0,3):
    print(Top3(z1)[i])
print()
for i in range(0,3):
    print(Top3(z2)[i])
print()
for i in range(0,3):
    print(Top3(z3)[i])
print()
for i in range(0,3):
    print(Top3(z4)[i])
print()
for i in range(0,3):
    print(Top3(z5)[i])
print()
```
