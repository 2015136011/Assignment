## 2)경로에 해당하는 문자열 1개를 입력 받아 그 안에 디렉토리 경로명과 파일명을 분리하여 리스트로 반환하는 함수      div_path(s)를 작성하시오.

-인자로 전달하는 문자열은 경로만 들어간다고 가정한다.

-각 디렉토리와 파일을 구분하는 문자는 '/'로 가정한다.

-반환하는 리스트의 첫번째 원소는 디렉토리이고 두번째 원소는 파일명이다.

-다음과 같은 실행 및 출력 결과가 도출되어야 한다.

    div_path('/usr/local/bin/python')
    
    ['/usr/local/bin', 'python']
    
    div_path('/home/chulsoo/test.txt')
    
    ['/home/chulsoo', 'test.txt']

```python
s=input('문자열을 입력하시오') #문자열을 입력을 받는다.
def div_path(s): #div_path(s)함수를 정의
    t=s.split('/') #'/'기준으로 문자열을 나누어 t에 넣는다.
    a=t[-1] #가장 마지막 리스트를 a에 넣는다.
    t.remove(a) #리스트t에서 가장 마지막 원소를 지운다.
    b='/'.join(t) #마지막 원소를 제외한 t 사이에 '/'를 추가하여 b에 넣는다.
    c="["+b+"', '" + a + "]" #b와 a를 다음 양식에 맞추어 c에 넣는다.
    return c #c를 반환

print(div_path(s)) #출력
```

먼저 문자열을 입력을 받았다. 이후에 함수를 정의하여 / 기준으로 문자열을 나누어 리스트 t에 넣었다. 최종 출력 때 마지막 원소를 따로 출력해야 하므로 미리 a에 넣어두고 t에서는 a를 제외한 원소들이 필요하므로 a를 지워주고 사이에 '/'를 입력하여 b에 넣고 출력 형태에 맞추어 b와 a 를 c에 넣어주고 최종적으로 출력을 한다.

## 3) 두 개의 리스트를 인자로 받아서 그 두 개의 리스트에 대한 '합집합'을 반환하는 함수 list_union(lista, listb)를 작성하시오.

-인자로 전달하는 리스트 2 개에는 정수값만 들어간다고 가정하자.

-함수 내에서 새로운 리스트를 만들어 그 리스트 내에 인자로 받은 두 리스트의 모든 원소를 넣어 반환한다. 

-반환하는 리스트에는 절대로 중복된 원소가 들어 있으면 안된다 (집합의 조건). 

-반환하는 리스트는 정렬이 되어 있어야 한다.

-다음과 같은 실행 및 출력 결과가 도출되어야 한다.

    list_union([1, 2, 3], [1, 2, 4]) 

    [1, 2, 3, 4] 

    list_union([-10, -5, 0, -1], [100, 9, 0, 9])

    [-10, -5, -1, 0, 9, 100] 

    list_union([0, 1, 2], [0, 1, 2]) 

    [0, 1, 2] 

[참고] 리스트(l) 내에 새로운 정수값 (예를 들어 10)을 넣는 방법은 l.append(10) 이다.

[참고] 임의의 정수값 (x)이 리스트 (l) 내에 존재하는지 판단하는 방법은 x in l 이다. 

```python
a=[-10,-10,-9,-10,-5,0,-1]
b=[100,9,0,9,8,9,0,0,0,0,0,0,0,0]
def list_union(a,b): #list_union(a,b) 함수를 정의
    c=a+b #a리스트와 b리스트를 합친다.
    for i in range (0,len(c)): #c의 크기만큼 반복
        x=c[0] #첫번째 요소를 x에 저장한다.
        c.remove(x) #x를 지운다.
        if(x in c):
            pass #x를 지워도 x와 같은 값이 리스트에 있다면 지운 상태로 넘어간다.
        else:
            c.append(x) #x를 지웠을 때 x랑 같은 값이 없을 경우 x는 하나인 것이므로 다시 x를 추가한다.
    c.sort() #c를 크기순서로 정렬한다.
    return c #c를 반환

print(list_union(a,b)) #출력
```

a와 b 리스트를 우선 정의한다. 함수 안에서 a와 b 리스트를 합쳐 c를 만든다. c의 첫번째 요소를 x에 저장 후에 삭제한다. 이후 첫번째 요소가 중복이 될 경우에는 삭제한 상태로 넘어가고 중복이 되지 않는다면 x를 다시 추가하여 준다. 첫번째 요소를 계속해서 반복하여 삭제하는 이유는 추가된 요소는 가장 마지막 요소로 들어가는 것이라 요소 순서대로 앞으로 당겨지기 때문이다. 이런식으로 중복된 값들을 제거하여주고 sort를 활용하여 숫자를 순서대로 정렬하여 출력한다.

# Incremental Project 문제

```python
import requests

req = requests.get('http://cse.koreatech.ac.kr')

source = req.text #컴공홈페이지의 소스를 받아온다.
```

```python
a="" #<>제외한 문자열을 넣어줄 a를 정의
i=0 #i를 0으로 정의
while (i < len(source)): #소스크기 만큼 반복하는 반복문
    if(source[i] == '<'): #요소가 <이 나올 경우
        i=i+1 #우선 다음 요소로 넘어가서
        while(source[i] != '>'): #요소가 >이 나올때까지
            i=i+1 #계속해서 다음 요소로 넘어간다.
        a=a+' ' # >이 나와서 while문을 탈출하면 공백을 a문자열에 넣어준다.
        i=i+1 # 다음 요소로 넘어감
    else:
        a=a+source[i] # 요소가 <가 아닐 경우에는 a문자열에 넣어준다.
        i=i+1 #다음 요소로 넘어감
b=a.split() #공백 단위로 a의 문자열을 나누어 b에 넣어준다. (단어를 구분함)
for j in range(0,len(b)): #b의 크기만큼
    print(b[j],end='\n') #단어들을 출력
print()
print()
print("총 단어의 개수는 {0}개입니다".format(len(b))) #단어개수 출력
```

<>제외한 문자열을 넣어줄 a를 가장 먼저 정의를 하고 반복문에 사용할 i도 0으로 정의해준다. 이후에 요소가 '<'이 나올 경우 '>' 이 나올때까지 다음 요소로 넘어가는 반복문을 만들어준다. >가 나왔을 경우에는 넘어가는 것을 그만 두고 공백을 띄어주는데 공백을 띄우는 이유는 예를 들어 가<나>다<라> 같은 경우에 <>만 삭제하면 '가'와 '다' 단어가 합쳐지기 때문에 > 이후에 공백을 넣어줌으로써 단어를 구분해준다. 요소가 '<'가 아닐 경우에는 해당 문자열을 a에 계속해서 넣어준다. 이후 a안에 있는 문자열들을 공백 단위로 토막내어 단어로 구분을 하여 b에 넣어주고 b에 단어들을 출력하며 최종적으로 b의 크기로 단어 개수를 출력한다.
