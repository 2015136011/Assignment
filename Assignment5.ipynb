## [3번 문제]

### 다음 각 요구사항 모두를 만족시키는 Counter 클래스를 코딩하시오 
#### (정답을 각 요구사항별로 입력할 필요 없이 3번 문제에 대해 1개의 클래스 정의 코드를 제시하면 된다.)

```python
class Counte():
    def __init__(self,count,step=1):
        self.count = count
        self.step = step
    def __str__(self):
        return '[Count (step: %d)] %d ' % (self.step, self.count)
    def incr(self):
        self.count = self.count + self.step
    def __call__(self):
        self.incr()
    def __add__(self, num):
        self.count += num
        return Counte(self.count)
    def __sub__(self, num):
        self.count -= num
        return Counte(self.count)
    def __lt__(self, other) :
        return self.count < other
    
    def __gt__(self, other) :
        return self.count > other
    
    def __eq__(self, other) :
        return self.count == other
```

```python
print('#1. 생성자 인자 받기')
c = Counte(10)
d = Counte(10, 2)
print('#2. 인스턴스 출력하는 __str__() 메소드로 c,d 출력')
print(c)
print(d)
print('#3.incr() 메소드를 통해 step만큼 count를 증가')
c.incr() 
d.incr() 
print(c)
print(d)
print('#4. 직접 호출하면서 그 안에 incr() 메소드 호출')
c()
d()
print(c)
print(d)
print('#5. 산술연산 메소드 추가')
c=c+5
d=d-5
print(c)
print(d)
print('#6. 관계연산 메소드 추가')
print (c > 10)
print (d > 10)
print (c < 10)
print (d < 10)
print (c == 17)
print (d != 9)
```

1.__init__()을 통해 count와 step을 생성하였다.  
2.__str__() 메소드를 통해 인스턴스 출력 양식을 과제에 맞게 반환하도록 하였다.  
3.incr() 메소드에서 self.count를 self.step만큼 더하게끔 하였다.  
4.call메소드 내에 incr() 메소드를 넣음으로써 자기자신 호출시 incr()하게끔 하였다.  
5.산술연산 메소드 add와 sub을 만들어 self.count에 other만큼 빼거나 더했다.  
6.관계연산 메소드 lt,gt,eq를 통하여 비교한 값이 맞는지를 True나 False로 반환하게끔 하였다.  

## [4번 문제]

### 다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서 __init__(), __str()__(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.

#### (MySet은 집합(Set) 자료형을 정의하려는 의도하에 만들어진 클래스이다.)

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
            
    #5-1 합집합 메소드
    def __or__(self,other):
        U=[]
        for e in self:
            U.append(e)
        for e in other:
            U.append(e)
        MySet.eliminate_duplicate(U)
        return MySet(U)
    
    #5-2 교집합 메소드
    def __and__(self, other):
        U = []
        for e1 in self:
            for e2 in other:
                if e1 == e2:
                    U.append(e2)
        return MySet(U)
    
    #5-3 차집합 메소드
    def __sub__(self, other):
        U=[]
        for e in self:
            if e not in other:
                U.append(e)  
        return MySet(U)

    
if __name__ == "__main__":
    s = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    t = MySet([1, 2, 2, 3,4,5])
    print(s)
    print(t)
    u = s | t
    print(u)
    u = s & t
    print(u)
    s = MySet([1, 2, 3])
    t = MySet([3, 4, 5])
    u = s - t
    print(u) 
 
```

###### init()
   self 리스트에 e의 원소들을 각각 넣어 self 리스트를 초기화한 후에 eliminate_duplicate()를 통하여 중복된 리스트의 원소들을 삭제한다.

###### str()
   result라는 문자열 "MySet: {"를 초기화한 후 리스트의 원소들을 각각 문자열의 형태로 result에 추가한다. 이후 result를 출력하는데 여기서result[0:len(result)-2]를 하는 이유는 마지막 쉼표와 공백을 지워주기 위해서이다.
   
###### eliminate_duplicate()
   리스트 s를 초기화한다. 입력된 리스트에 각각의 원소가 s리스트에 없다면 해당 원소를 s 리스트에 추가시켜준다.입력된 리스트인 l을 비어있는 리스트로 초기화를 한 후. 다시 중복값을 제외한 리스트인 e리스트의 원소들을 넣어준다.

8) searchUrlByword(word) : 앞서 말했듯이 url별로 원소를 리스트화 한 함수 getWords를 불러온다. 이후 s라는 리스트를 정의하여 각 url별로 사전화하여 리스트 s에 넣어준다. 유사도를 비교하기 위한 유사도가 0이 아닌 url들의 리스트인 USA와 최고 유사도 값이 maxUsa를 정의하여주고 word가 s의 각 리스트 원소 안에 있다하면 정의되있던 maxUsa보다 클 경우 그 리스트에서 word의 value값을 maxUsa로 하고 USA 리스트에 해당 url을 삽입해준다. 이렇게 모든 url들을 반복하여 얻어진 maxUsa 값이 처음 정의햇던 그대로 0이라면 모든 주소가 유사도가 0인 것이기에 None을 출력해주고 아니라면 USA 리스트에 각 원소의 value값이 최종적인 유사값 maxUsa와 같으면 해당 url을 출력을 하여준다.
